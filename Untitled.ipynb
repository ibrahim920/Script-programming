```python
##What's the average overall for each club (based on the players they have on the list)?
results = players.groupby(["Club"]).mean(["Overall"])
print(results.head(5))
```

### The aggregate age of all players in each team.

```python
results = players.groupby('Club', as_index=False).agg({'Age': 'sum'})
print(results.head(20))
```
